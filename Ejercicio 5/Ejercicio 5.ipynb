```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_convolution_filter(image_path, kernel_type, kernel_size):
    # Cargar la imagen a color
    color_image = cv2.imread(image_path)
    
    # Definir los kernels según el tipo seleccionado
    if kernel_type == "gaussian":
        kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size * kernel_size)
    elif kernel_type == "sobel":
        kernel = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
    elif kernel_type == "sharp":
        kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    elif kernel_type == "unsharp":
        kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
    elif kernel_type == "laplacian":
        kernel = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])
    else:
        raise ValueError("Tipo de kernel no válido")
    
    # Aplicar la convolución a cada canal de la imagen
    filtered_image = cv2.filter2D(color_image, -1, kernel)
    
    # Mostrar la imagen original y la filtrada
    plt.figure(figsize=(10, 6))
    
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB))
    plt.title("Imagen Original")
    
    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(filtered_image, cv2.COLOR_BGR2RGB))
    plt.title(f"Imagen Filtrada ({kernel_type.capitalize()}, {kernel_size}x{kernel_size})")
    
    plt.tight_layout()
    plt.show()

# Ruta de la imagen a color
input_image_path = "./2021-03-05-18-45-01_0.png"

# Parámetros del filtro de convolución
kernel_type = "gaussian"  # Cambiar al tipo de filtro deseado: "gaussian", "sobel", "sharp", "unsharp", "laplacian"
kernel_size = 5  # Cambiar al tamaño de kernel deseado: 3, 5, 7, 9, 11

# Aplicar el filtro de convolución y mostrar las imágenes
apply_convolution_filter(input_image_path, kernel_type, kernel_size)
```
